```python
class NonAssociativeAlgebra:
    
    def __init__(self, n,parameters):
        self.n=n
        if len(parameters)==1:
            self.listBrackets=parameters[0]
        if len(parameters)==2:
            self.listBrackets=parameters[0]
            self.metric=parameters[1]
        for i in [1..n]:
            for j in [1..n]:
                parameters[0][(i,j)] = vector(SR,n)
            
            
    def coordinates(self,i):
        v=vector(SR,self.n)
        v[i-1]=1
        return v 

    def nonVanishingBracket(self,i,j,u):
        self.listBrackets.update({(i,j):u})
        


    def bracket(self,u,v):
        s = vector(SR,self.n)
        for i in [0..self.n-1]:
            for j in [0..self.n-1]:
                s+=  u[i]*v[j]*self.listBrackets[(i+1,j+1)]
        return s
    
    def leftLeibniz(self,u,v,w):
        return self.bracket(w,self.bracket(u,v))-self.bracket(self.bracket(w,u),v)-self.bracket(u,self.bracket(w,v))

    def leftLeibnizAll(self):
        dict = {}
        for i in [1..self.n]:
            for j in [1..self.n]:
                for k in [1..self.n]:
                    dict[(i,j,k)]=self.leftLeibniz(self.coordinates(i),self.coordinates(j),self.coordinates(k))
        return dict;
    
    def rightLeibniz(self,u,v,w):
        return self.bracket(self.bracket(u,v),w)-self.bracket(self.bracket(u,w),v)-self.bracket(u,self.bracket(v,w))

    def rightLeibnizAll(self):
        dict = {}
        for i in [1..self.n]:
            for j in [1..self.n]:
                for k in [1..self.n]:
                    dict[(i,j,k)]=self.rightLeibniz(self.coordinates(i),self.coordinates(j),self.coordinates(k))
        return dict;
    
    
    def leftMultiplication(self,u):
       
        M = matrix(SR,self.n,self.n)
        for i in [0..self.n-1]:
            for j in [0..self.n-1]:
                M[i,j] = self.bracket(u,self.coordinates(j+1))[i]
        return M 

    def rightMultiplication(self,u):
        
        M = matrix(SR,self.n,self.n)
        for i in [0..self.n-1]:
            for j in [0..self.n-1]:
                M[i,j] = self.bracket(self.coordinates(j+1),u)[i]
        return M
    
    def adjoint(self,u):
        return 1/2*(self.leftMultiplication(u)-self.rightMultiplication(u))
    
    def symmetricAdjoint(self,u):
        return 1/2*(self.leftMultiplication(u)+self.rightMultiplication(u))
                    

    def killing(self,u,v):
        return -1/2*(self.leftMultiplication(u)*self.rightMultiplication(v)+\
                     self.rightMultiplication(u)*self.leftMultiplication(v)).trace()
  

    def killingMatrix(self):
        return matrix(SR,self.n,self.n, lambda i,j:self.killing(self.coordinates(i+1),self.coordinates(j+1)) )

    def myMatrix(self):
        plist = ['p%s%s' %(i,j) for i in range(self.n) for j in range(self.n)]
        return matrix(SR,self.n,self.n,plist)


    def skewSymmetricMatrix(self):
        plist = ['p%s%s' %(i,j) for i in range(self.n) for j in range(self.n)]
        M=matrix(SR,self.n,self.n,plist)
        A = copy(M) 
        for i in [0..self.n-1]:
            for j in [i..self.n-1]:
                A[i,i]=0
                A[j,i]=-A[i,j]
        return A     

    def symmetricMatrix(self):
        plist = ['p%s%s' %(i,j) for i in range(self.n) for j in range(self.n)]
        M=matrix(SR,self.n,self.n,plist)
        A = copy(M) 
        for i in [0..self.n-1]:
            for j in [i..self.n-1]:
                A[j,i]=A[i,j]
        return A
    
    def plist(self):
        K=[]
        n=self.n
        for i in [0..n-1]:
            for j in [0..n-1]:
                var('p'+str(i)+str(j))
                K.append(var('p'+str(i)+str(j)))
        return K        
        

    

    def derivation(self):
        n=self.n
        M=self.myMatrix()
        L=[]
        K=[]
        for i in [0..n-1]:
            for j in [0..n-1]:
                var('p'+str(i)+str(j))
                K.append(var('p'+str(i)+str(j)))
        for i in [1..n]:
            for j in [1..n]:
                for k in [0..n-1]:
                    L.append((M*self.bracket(self.coordinates(i),self.coordinates(j)))[k]\
                             -self.bracket(M*self.coordinates(i),self.coordinates(j))[k]-\
                             self.bracket(self.coordinates(i),M*self.coordinates(j))[k]==0)
        
        return  M.substitute(solve(L,K))

    def automorphism(self):
        n=self.n
        M=self.myMatrix()
        L=[]
        K=[]
        for i in [0..n-1]:
            for j in [0..n-1]:
                var('p'+str(i)+str(j))
                K.append(var('p'+str(i)+str(j)))
        for i in [1..n]:
            for j in [1..n]:
                for k in [0..n-1]:
                    L.append((M*self.bracket(self.coordinates(i),self.coordinates(j)))[k]\
                             -self.bracket(M*self.coordinates(i),M*self.coordinates(j))[k]==0)
        m=len(solve(L,K))
        S=[]
        for i in [0..m-1]:
            if (M.substitute(solve(L,K)[i])).determinant()!=0:
                S.append(M.substitute(solve(L,K)[i]))
        return  S 
    
    def isAutomorphism(self,A):
        n= self.n
        dict ={}
        for i in [1..n]:
            for j in [1..n]:
                dict[(i,j)]=A*self.bracket(self.coordinates(i),self.coordinates(j))\
                -self.bracket(A*self.coordinates(i),A*self.coordinates(j))
        return dict     
    
    def solution(self,M,L):
        n=len(L)
        l=M.nrows()
        c=M.ncols()
        K=[]
        for i in [0..n-1]:
            var(L[i])
        for i in [0..l-1]:
            for j in [0..c-1]:
                K.append(M[i,j]==0)
        return solve(K,L)
    
    def scalarProduct(self,u,v):
        return u*self.metric*v
    
    def metricTranspose(self,A):
        return self.metric.inverse()*(A.transpose())*self.metric
    
    
    def leftMeanVector(self):
        n=self.n
        X=vector(SR,n)
        for i in [0..n-1]:
            X[i]=self.leftMultiplication(self.coordinates(i+1)).trace()
        return self.metric.inverse()*X
    
    def rightMeanVector(self):
        n=self.n
        X=vector(SR,n)
        for i in [0..n-1]:
            X[i]=self.rightMultiplication(self.coordinates(i+1)).trace()
        return self.metric.inverse()*X 
    
    def leftStructureEndomorphism(self,u):
        n=self.n
        M=matrix(SR,n,n)
        for i in [0..self.n-1]:
            for j in [0..self.n-1]:
                M[i,j] = (self.metricTranspose(self.leftMultiplication(self.coordinates(j+1)))*u)[i]      
        return M
    
    def rightStructureEndomorphism(self,u):
        n=self.n
        M=matrix(SR,n,n)
        for i in [0..self.n-1]:
            for j in [0..self.n-1]:
                M[i,j] = (self.metricTranspose(self.rightMultiplication(self.coordinates(j+1)))*u)[i]
        return M
    
    def ajointStructureEndomorphism(self,u):
        return 1/2*(self.leftStructureEndomorphism(u)-self.rightStructureEndomorphism(u))
    
    
    def symmetricStructureEndomorphism(self,u):
        return 1/2*(self.leftStructureEndomorphism(u)+self.rightStructureEndomorphism(u))
    
    
    def leftRicci(self,u,v):
        return -1/2*(self.leftMultiplication(u)*self.leftMultiplication(v)).trace()\
               -1/2*(self.leftMultiplication(u)*self.metricTranspose(self.leftMultiplication(v))).trace()\
               -1/4*(self.leftStructureEndomorphism(u)*self.leftStructureEndomorphism(v)).trace()\
               -1/2*self.scalarProduct(self.bracket(self.leftMeanVector(),u),v)\
               -1/2*self.scalarProduct(self.bracket(self.leftMeanVector(),v),u)
    
    def rightRicci(self,u,v):
        return -1/2*(self.rightMultiplication(u)*self.rightMultiplication(v)).trace()\
               -1/2*(self.rightMultiplication(u)*self.metricTranspose(self.rightMultiplication(v))).trace()\
               -1/4*(self.rightStructureEndomorphism(u)*self.rightStructureEndomorphism(v)).trace()\
               -1/2*self.scalarProduct(self.bracket(u,self.rightMeanVector()),v)\
               -1/2*self.scalarProduct(self.bracket(v,self.rightMeanVector()),u)
    
    def matrixLeftRicci(self):
        n=self.n
        return matrix(SR,n,n, lambda i,j:self.leftRicci(self.coordinates(i+1),self.coordinates(j+1)))
 
    def leftRicciTensor(self):
        return self.metric.inverse()*self.matrixLeftRicci()
    
    def matrixRightRicci(self):
        n=self.n
        return matrix(SR,n,n, lambda i,j:self.rightRicci(self.coordinates(i+1),self.coordinates(j+1)))
 
    def rightRicciTensor(self):
        return self.metric.inverse()*self.matrixRightRicci()
    
    def symmetricForm(self,u,v):
        return (self.symmetricStructureEndomorphism(u)\
                *self.symmetricStructureEndomorphism(v)).trace()\
               -(self.symmetricAdjoint(u)*\
                 self.metricTranspose(self.symmetricAdjoint(v))).trace()
    
    def matrixSymmetricForm(self):
        n=self.n
        return matrix(SR,n,n, lambda i,j:self.symmetricForm(self.coordinates(i+1),self.coordinates(j+1)))
 
    def symmetricFormTensor(self):
        return self.metric.inverse()*self.matrixSymmetricForm()
    
              
    
    def ricciTensor(self):
        return 1/2*(self.leftRicciTensor()+self.rightRicciTensor())+self.symmetricFormTensor()
    
    def matrixRicci(self):
        return 1/2*(self.matrixLeftRicci()+self.matrixRightRicci())+self.matrixSymmetricForm()
               
 
     
    

    
    
    
```
