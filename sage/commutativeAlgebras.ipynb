```python
class CommutativeAlgebra:
    
    def __init__(self, n,parameters):
        self.n=n
        if len(parameters)==1:
            self.listBrackets=parameters[0]
        if len(parameters)==2:
            self.listBrackets=parameters[0]
            self.metric=parameters[1]
        for i in [1..n]:
            for j in [1..n]:
                parameters[0][(i,j)] = vector(SR,n)
            
            
                
    def coordinates(self,i):
        v=vector(SR,self.n)
        v[i-1]=1
        return v 

    def nonVanishingBracket(self,i,j,u):
        self.listBrackets.update({(i,j):u})
        self.listBrackets.update({(j,i):u})


    def bracket(self,u,v):
        s = vector(SR,self.n)
        for i in [0..self.n-1]:
            for j in [0..self.n-1]:
                s+=  u[i]*v[j]*self.listBrackets[(i+1,j+1)]
        return s
    
    def associator(self,u,v,w):
        return self.bracket(self.bracket(u,v),w)-self.bracket(u,self.bracket(v,w))

    def associatorAll(self):
        dict = {}
        for i in [1..self.n]:
            for j in [1..self.n]:
                for k in [1..self.n]:
                    dict[(i,j,k)]=self.associator(self.coordinates(i),self.coordinates(j),self.coordinates(k))
        return dict;
    
    def adjoint(self,u):
       
        M = matrix(SR,self.n,self.n)
        for i in [0..self.n-1]:
            for j in [0..self.n-1]:
                M[i,j] = self.bracket(u,self.coordinates(j+1))[i]
        return M 

    
    def killing(self,u,v):
        return (self.adjoint(u)*self.adjoint(v)).trace()

    def killingMatrix(self):
        return matrix(SR,self.n,self.n, lambda i,j:self.killing(self.coordinates(i+1),self.coordinates(j+1)) )

    def myMatrix(self):
        plist = ['p%s%s' %(i,j) for i in range(self.n) for j in range(self.n)]
        return matrix(SR,self.n,self.n,plist)


    def skewSymmetricMatrix(self):
        plist = ['p%s%s' %(i,j) for i in range(self.n) for j in range(self.n)]
        M=matrix(SR,self.n,self.n,plist)
        A = copy(M) 
        for i in [0..self.n-1]:
            for j in [i..self.n-1]:
                A[i,i]=0
                A[j,i]=-A[i,j]
        return A     

    def symmetricMatrix(self):
        plist = ['p%s%s' %(i,j) for i in range(self.n) for j in range(self.n)]
        M=matrix(SR,self.n,self.n,plist)
        A = copy(M) 
        for i in [0..self.n-1]:
            for j in [i..self.n-1]:
                A[j,i]=A[i,j]
        return A 

    
    def derivation(self):
        n=self.n
        M=self.myMatrix()
        L=[]
        K=[]
        for i in [0..n-1]:
            for j in [0..n-1]:
                var('p'+str(i)+str(j))
                K.append(var('p'+str(i)+str(j)))
        for i in [1..n]:
            for j in [i..n]:
                for k in [0..n-1]:
                    L.append((M*self.bracket(self.coordinates(i),self.coordinates(j)))[k]\
                             -self.bracket(M*self.coordinates(i),self.coordinates(j))[k]-\
                             self.bracket(self.coordinates(i),M*self.coordinates(j))[k]==0)
        
        return  M.substitute(solve(L,K))

    def automorphism(self):
        n=self.n
        M=self.myMatrix()
        L=[]
        K=[]
        for i in [0..n-1]:
            for j in [0..n-1]:
                var('p'+str(i)+str(j))
                K.append(var('p'+str(i)+str(j)))
        for i in [1..n]:
            for j in [i..n]:
                for k in [0..n-1]:
                    L.append((M*self.bracket(self.coordinates(i),self.coordinates(j)))[k]\
                             -self.bracket(M*self.coordinates(i),M*self.coordinates(j))[k]==0)
        m=len(solve(L,K))
        S=[]
        for i in [0..m-1]:
            if (M.substitute(solve(L,K)[i])).determinant()!=0:
                S.append(M.substitute(solve(L,K)[i]))
        return  S 
    
    def isAutomorphism(self,A):
        n= self.n
        dict ={}
        for i in [1..n]:
            for j in [1..n]:
                dict[(i,j)]=A*self.bracket(self.coordinates(i),self.coordinates(j))\
                -self.bracket(A*self.coordinates(i),A*self.coordinates(j))
        return dict        
    
    def scalarProduct(self,u,v):
        return u*self.metric*v
    
    def metricTranspose(self,A):
        return self.metric.inverse()*(A.transpose())*self.metric
    
    def meanVector(self):
        n=self.n
        X=vector(SR,n)
        for i in [0..n-1]:
            X[i]=self.adjoint(self.coordinates(i+1)).trace()
        return self.metric.inverse()*X
    
    def meanBracketVector(self):
        n=self.n
        X=vector(SR,n)
        for i in [0..n-1]:
            X[i]=self.structureEndomorphism(self.coordinates(i+1)).trace()
        return self.metric.inverse()*X
    
    def structureEndomorphism(self,u):
        n=self.n
        M=matrix(SR,n,n)
        for i in [0..self.n-1]:
            for j in [0..self.n-1]:
                M[i,j] = (self.metricTranspose(self.adjoint(self.coordinates(j+1)))*u)[i]      
        return M

    
    
    
    def leviCivitaProduct(self,u,v):
        n = self.n
        X = vector(SR,n)
        for i in [0..n-1]:
            X[i] = 1/2*self.scalarProduct(self.bracket(u,v),self.coordinates(i+1))\
            +1/2*self.scalarProduct(self.bracket(self.coordinates(i+1),u),v)\
            -1/2*self.scalarProduct(self.bracket(self.coordinates(i+1),v),u)
        return (self.metric.inverse())*X
 
    def leftLeviCivitaProduct(self,u):
        n = self.n
        M = matrix(SR,n,n)
        for i in [0..n-1]:
            for j in [0..n-1]:
                M[i,j] = self.leviCivitaProduct(u,self.coordinates(j+1))[i]
        return M 
    
    def rightLeviCivitaProduct(self,u):
        n = self.n
        M = matrix(SR,n,n)
        for i in [0..n-1]:
            for j in [0..n-1]:
                M[i,j] = self.leviCivitaProduct(self.coordinates(j+1),u)[i]
        return M

    def ricci(self,u,v):
        r= (self.rightLeviCivitaProduct(u)*self.rightLeviCivitaProduct(v)).trace()\
        -self.rightLeviCivitaProduct(self.leviCivitaProduct(u,v)).trace()
        return r

    def matrixRicci(self):
        n=self.n
        return matrix(SR,n,n, lambda i,j:self.ricci(self.coordinates(i+1),self.coordinates(j+1)))
 
    def ricciTensor(self):
        return self.metric.inverse()*self.matrixRicci()
 
    def curvature(self,u,v):
        return self.leftLeviCivitaProduct(self.bracket(u,v))-\
        self.leftLeviCivitaProduct(u)*self.leftLeviCivitaProduct(v)\
        -self.leftLeviCivitaProduct(v)*self.leftLeviCivitaProduct(u)
 
    def curvatureAll(self):
        n=self.n
        dict = {}
        for i in [1..n]:
            for j in [i..n]:
                dict[(i,j)] = self.curvature(self.coordinates(i),self.coordinates(j))
        return dict 
  
    def solution(self,M,L):
        n=len(L)
        l=M.nrows()
        c=M.ncols()
        K=[]
        for i in [0..n-1]:
            var(L[i])
        for i in [0..l-1]:
            for j in [0..c-1]:
                K.append(M[i,j]==0)
        return solve(K,L)
    
    def plist(self):
        K=[]
        n=self.n
        for i in [0..n-1]:
            for j in [0..n-1]:
                var('p'+str(i)+str(j))
                K.append(var('p'+str(i)+str(j)))
        return K
    








```
